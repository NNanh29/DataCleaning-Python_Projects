```python
import pandas as pd
ride_sharing=pd.read_csv(r'C:\Users\Ngoc Anh\Downloads\data_ride_sharing_updated.csv')
```

```python
ride_sharing.head()
```

```python
#Find duplicates values in ride_id
subset={'ride_id'}
duplicates=ride_sharing.duplicated(subset=subset, keep=False)
```

```python
ride_sharing[duplicates].sort_values(by='ride_id')
```

```python
#Drop complete duplicated values
ride_sharing.drop_duplicates(inplace=True)
```

```python
#Replace incomplelte values by mean
ride_sharing.info()
ride_sharing.head()
statistic={'duration':'mean'}
ride_sharing=ride_sharing.groupby(by='ride_id').agg(statistic).reset_index()
```
